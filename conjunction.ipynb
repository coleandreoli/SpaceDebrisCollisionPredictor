```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, make_scorer, fbeta_score, accuracy_score

# Assume 'df' is your DataFrame and 'oe' is a list of feature columns
X = df[oe]          # Feature columns (orbital parameters)
y = df['collision'] # Target column (True if collision, False otherwise)

# Split data with stratification to maintain class distribution
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# # Apply SMOTE to balance the training data
# smote = SMOTE(random_state=42)
# X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# Define a custom scorer prioritizing recall using an F-beta score with beta=2.
fbeta_scorer = make_scorer(fbeta_score, beta=2)

# Define a smaller parameter grid (search space)
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Initialize the Random Forest with custom class weights to favor the True class.
rf = RandomForestClassifier(class_weight={False: 1, True: 20}, random_state=42)

# Use RandomizedSearchCV for a faster search:
random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_grid,
    n_iter=10,       # Try 10 random combinations instead of all combinations
    scoring=fbeta_scorer,
    cv=5,            # Use n-fold cross-validation
    n_jobs=-1,       # Use all available cores
    verbose=1,
    random_state=42
)

# Run the random search on the resampled training data
# random_search.fit(X_train_res, y_train_res)
random_search.fit(X_train, y_train)
print("Best Parameters:", random_search.best_params_)

# Use the best estimator from the random search for predictions
best_clf = random_search.best_estimator_
# If you want to adjust the decision threshold to further favor the True class:
y_probs = best_clf.predict_proba(X_test)[:, 1]
```

```python
import pickle
SAVE = False
if SAVE:
    # Save model
    with open("random_forest_model.pkl", "wb") as f:
        pickle.dump(best_clf, f)
```

```python
threshold = .3  # Adjust threshold as needed to favor the True class
y_pred_custom = np.where(y_probs > threshold, True, False)

# Evaluate the model
print("Classification Report with Custom Threshold:")
print(classification_report(y_test, y_pred_custom))
print("Accuracy:", accuracy_score(y_test, y_pred_custom))

clf = best_clf
y_pred = y_pred_custom
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Generate the confusion matrix from test data and predictions
cm = confusion_matrix(y_test, y_pred)

# Create a heatmap using seaborn
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Collision', 'Collision'],
            yticklabels=['No Collision', 'Collision'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python

# Get feature importances
feature_importance = clf.feature_importances_
features = X.columns

# Create a DataFrame for visualization
importance_df = pd.DataFrame({"Feature": features, "Importance": feature_importance})
importance_df = importance_df.sort_values(by="Importance", ascending=False)

# Plot feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x="Importance", y="Feature", data=importance_df)
plt.xlabel("Feature Importance Score")
plt.ylabel("Features")
plt.title("Feature Importance from Random Forest")
plt.show()
```
